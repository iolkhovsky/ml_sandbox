```python
from pca_impl.pca_computer import PCAComputer
from pca_impl.utils import generate_correlated_data
from pca_impl.visualization import visualize_2d_pca, visualize_classification_2d


X = generate_correlated_data()
y = [0] * len(X)
```

```python
visualize_classification_2d(X, y)
```

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=2).fit(X)
visualize_2d_pca(X, pca.components_, pca.explained_variance_)
```

```python
pca = PCAComputer(2)
pca.fit(X)
transformed = pca(X)

visualize_classification_2d(transformed, [0] * len(transformed))
visualize_2d_pca(X, pca.principal_components_, pca.eigen_values_)
```
